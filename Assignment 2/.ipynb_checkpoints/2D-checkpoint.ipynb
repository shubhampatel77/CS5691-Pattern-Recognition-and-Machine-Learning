```python
import  numpy as np
import matplotlib.pyplot as plt
```

```python
fname = '2d-team-7-train.txt'
```

```python
xone,xtwo,t = np.loadtxt(fname, skiprows=1, unpack=True)

# ax = plt.axes(projection ='3d')
# ax.scatter3D(xone,xtwo,t)
# plt.show()

# fig = plt.figure(figsize =(14, 9))
# ax = plt.axes(projection ='3d')
 
# # Creating plot
# ax.plot_surface(xone, xtwo, t)
 
# # show plot
# plt.show()
```

```python
def makephi(n,m):
    phi = np.ones(999)
    done = np.arange(1,n,1)
    dtwo = np.arange(1,m,1)
    for i in done:
        phi = np.column_stack((phi,xone**i))
        
        
    for j in dtwo:
        phi = np.column_stack((phi,xtwo**j))
        
    return phi
# phi = np.column_stack((phi,xone))
# phi = np.column_stack((phi,xtwo))
# phi = np.column_stack((phi,xtwo**2))
# phi = np.column_stack((phi,xone**2))

phi = makephi(20,20)
phi.shape


```

```python
rang = np.arange(2,25,1)
summ = 100
ind = [1,1]
for i in rang:
    for j in rang:
        phi = makephi(i,j)
        r = phi.T@phi
        l = np.linalg.inv(r)
        fin = l@phi.T@t
        final = phi@fin
        deter = (t-final)**2
        sume = 0
        for k in deter:
            sume = sume + k
        sume
        if sume<summ:
            summ = sume
            ind = [i,j]  
ind


```

```python
phi = makephi(23,20)
r = phi.T@phi + np.identity(42)
l = np.linalg.inv(r)
fin = l@phi.T@t
final = phi@fin
phi.shape
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(xone,xtwo,t,color='green')
 
# Creating plot
surf = ax.plot_trisurf(xone, xtwo, final,color='gray')
ax.view_init(10, 90)
# show plot
plt.show()
```
