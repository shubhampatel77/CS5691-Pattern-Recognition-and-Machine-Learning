```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy.optimize import fsolve
```

```python
train = np.loadtxt('train.txt', delimiter =',')
dev = np.loadtxt('dev.txt', delimiter =',')
```

```python
def trainclass(x):
    c1 = x[:350, :]; c2 = x[350:700, :]; c3 = x[700:1050, :]
    return (c1, c2, c3)
```

```python
def devclass(x):
    c1 = x[:100, :]; c2 = x[100:200, :]; c3 = x[200:300, :]
    return (c1, c2, c3)
```

```python
def stats(x):
    μ = np.array([np.mean(x[:,0]), np.mean(x[:,1])])
    μ = np.reshape(μ, (len(μ),1)).T
    Σ = np.cov(x, rowvar=False)
    return (μ, Σ)
```

```python
def gaussian(x, μ, Σ):
    d = len(μ[0])
    exponent = np.exp(-0.5*(x-μ)@np.linalg.inv(Σ)@(x-μ).T )
    det = 1/(np.linalg.det(Σ))**0.5
    gdf = 1/(2*np.pi)**(d/2)*det*exponent
    return float(gdf[0])
```

```python
def bayesian(x, xtest):
    (c1, c2, c3) = trainclass(x)
    (m1, s1) = stats(c1[:, :2]); (m2, s2) = stats(c2[:, :2]); (m3, s3) = stats(c3[:, :2])
    #c = np.cov(x[:, :2], rowvar=False) for same cov
    ytest = np.empty([0],dtype=float)
    for i in range(len(xtest)):
        gdf1 = gaussian(xtest[i,:], m1, s1)#here
        gdf2 = gaussian(xtest[i,:], m2, s2)#here
        gdf3 = gaussian(xtest[i,:], m3, s3)#here
        ytest = np.append(ytest,np.argmax([gdf1,gdf2,gdf3])+1)
    return ytest
```

```python
def ROC(x, xtest, ytest):
    (c1, c2, c3) = trainclass(x)
    (m1, s1) = stats(c1[:, :2]); (m2, s2) = stats(c2[:, :2]); (m3, s3) = stats(c3[:, :2])
    m = np.concatenate([m1, m2, m3], axis=0)
    c = np.cov(x[:, :2], rowvar=False)#same 
    #s = np.dstack((s1,s2,s3)) for diff
    
    threshold_min = 1
    threshold_max = 0
    for i in range(len(xtest)):
        p1 = 1/3*gaussian(xtest[i,:], m1, s1)
        p2 = 1/3*gaussian(xtest[i,:], m2, s2)
        p3 = 1/3*gaussian(xtest[i,:], m3, s3)
        p = np.array([p1, p2, p3])
        if np.min(p)<=threshold_min:
            threshold_min = np.min(p)
        elif np.max(p)>=threshold_max:
            threshold_max = np.max(p)
   
    threshold_vec = np.linspace(threshold_min, threshold_max, 100)
    P = np.empty(3)
    TPR = []; FPR = []
    for threshold in threshold_vec:
        TP, FP = 0, 0
        TN, FN = 0, 0
        for i in range(len(xtest)):
            ypredict = ytest[i]
            for j in range(3):
                P[j] = 1/3*gaussian(xtest[i,:], np.array([m[j, :]]), c)#cov term change here
                if P[j]>=threshold:
                    if ypredict == j+1:
                        TP += 1
                    else:
                        FP += 1
                else:
                    if ypredict == j+1:
                        FN += 1
                    else:
                        TN += 1
        TPR.append(TP/(TP+FN))
        FPR.append(FP/(FP+TN))
    return(TPR, FPR)   
```

```python
(c1, c2, c3) = trainclass(train)
(m1, s1) = stats(c1[:, :2]); (m2, s2) = stats(c2[:, :2]); (m3, s3) = stats(c3[:, :2])

min1, max1 = train[:, 0].min()-1, train[:, 0].max()+1
min2, max2 = train[:, 1].min()-1, train[:, 1].max()+1
x1grid = np.arange(min1, max1, 0.5)
x2grid = np.arange(min2, max2, 0.5)
xx, yy = np.meshgrid(x1grid, x2grid)
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
grid = np.hstack((r1,r2))

gdf1 = []; gdf2 = []; gdf3 = []
for i in range(len(grid)):
    g1 = gaussian(grid[i,:], m1, s1); gdf1.append(g1)
    g2 = gaussian(grid[i,:], m2, s2); gdf2.append(g2)
    g3 = gaussian(grid[i,:], m3, s3); gdf3.append(g3)

y = bayesian(train, grid)
z = np.reshape(y,(((np.shape(x2grid))[0]),((np.shape(x1grid))[0])))
plt.title("Bayesian with Different Covariance for all classes")

plt.contourf(xx, yy, z, levels=2)
plt.scatter(c1[:, 0], c1[:, 1],marker="*",)
plt.scatter(c2[:, 0], c2[:, 1],marker="x")
plt.scatter(c3[:, 0], c3[:, 1],marker=".")
plt.savefig("D1C1.jpg")
plt.show()
```

```python
(TPR, FPR) = ROC(train, grid, y)
plt.plot(FPR, TPR)
```
